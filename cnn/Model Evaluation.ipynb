```python
import pickle
import matplotlib.pyplot as plt
with open("/home/yiqin/2018summer_project/DeepMusic/cnn/cnn_val_crossentropy_augmented.pk", "rb") as f:
    precision = pickle.load(f)
    recall = pickle.load(f)
    losses = pickle.load(f)

for l in losses:
    plt.plot(l)
plt.show()
```

```python
import numpy as np
x = np.arange(len(precision))
plt.plot(x, precision)
plt.plot(x, recall)
plt.legend(["precision", "recall"], loc = "upper right")
plt.show()
```

```python
precision=np.array(precision)
recall=np.array(recall)
F1 = 2.0/((1.0/precision)+(1.0/recall))
plt.plot(x, F1)
plt.show()
```

```python
F1=F1.reshape(5,-1)
max_F1=F1.max(axis=1)
precision=precision.reshape(5,-1)
max_precision=precision.max(axis=1)
recall=recall.reshape(5,-1)
max_recall=recall.max(axis=1)
print(max_F1)
mean_F1=max_F1.sum()/5
print(mean_F1)
print((np.sum(((max_F1-mean_F1)*100)**2)/5)**0.5)
```

```python
print(max_precision)
mean_p=max_precision.sum()/5
print(mean_p)
print((np.sum(((max_precision-mean_p)*100)**2)/5)**0.5)
```
