```python
import pickle
import matplotlib.pyplot as plt
with open("/home/yiqin/2018summer_project/DeepMusic/cnn/cnn_val_smooth_augmented.pk", "rb") as f:
    precision = pickle.load(f)
    recall = pickle.load(f)
    losses = pickle.load(f)

for l in losses:
    plt.plot(l)
plt.show()
```

```python
import numpy as np
x = np.arange(len(precision))
plt.plot(x, precision)
plt.plot(x, recall)
plt.legend(["precision", "recall"], loc = "upper right")
plt.show()
```
