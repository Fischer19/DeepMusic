```python
import math
def compute_F1(val_recall, val_precision, epoch = 1, partition = 5):
    temp = 0
    temp1 = 0
    temp2 = 0
    std1 = 0
    std2 = 0
    std3 = 0
    temp3 = []
    best = []
    for i in range(partition):
        best_pair = (0,0)
        best_score = 0
        for j in range(16 * epoch):
            iter = i*16*epoch + j
            precision = float(val_precision[iter])
            recall = float(val_recall[iter])
            if (precision+recall) == 0:
                pass
            elif 2*(precision*recall)/(precision+recall) > best_score:
                best_score = 2*(precision*recall)/(precision+recall)
                best_pair = (precision, recall)
        print(best_score)
        best.append(best_score)
        temp+=best_score
        temp1+=best_pair[0]
        temp2+=best_pair[1]
        temp3.append(best_pair)
    print(temp3)
    for i in range(partition):
        std1+=(temp3[i][0] - temp1/5)**2
        std2+=(temp3[i][1] - temp2/5)**2
        std3+=(best[i] - temp/5)**2
    return {"precision": temp1/partition, "recall" : temp2/partition, "F1": temp/partition, 
            "precision std":math.sqrt(std1/partition), "recall std": math.sqrt(std2/partition), "F1 std":math.sqrt(std3/partition)}
```
